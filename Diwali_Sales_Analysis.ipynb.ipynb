```python
# --- Step 1 (RELOAD) ---
# --- Step 2 (DROP EMPTY COLUMNS) ---
# --- Step 3 (DROP MISSING VALUES) ---
# --- Step 4 (FIX DATA TYPE) ---

df = pd.read_csv('Diwali Sales Data.csv', encoding='ISO-8859-1')

df.drop(['Status', 'unnamed1'], axis=1, inplace=True)

df.dropna(inplace=True)

df['Amount'] = df['Amount'].astype('int64')

print("--- Final Cleaned Data Info ---")
df.info()

print("\n--- Final Cleaned Data Head ---")
print(df.head())
```

```python
print(df.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(7,5)})
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
# --- Analysis: Total Sales Amount by Gender ---

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
ax = sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)

plt.show()
```

```python
# --- Analysis: Customer Count by Age Group and Gender ---

sns.set(rc={'figure.figsize':(12,7)})
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender', order = sorted(df['Age Group'].unique()))

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
# --- Analysis: Total Sales Amount by Age Group ---

sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(12,7)}) # Make the chart wide
ax = sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age, order = sales_age['Age Group'])

plt.show()
```

```python
# --- Analysis: Top 10 States by Total Sales ---

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,7)})
ax = sns.barplot(data = sales_state, x = 'State', y = 'Amount')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
# --- Analysis: Top 10 Product Categories by Sales ---

sales_product = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,7)})
ax = sns.barplot(data = sales_product, x = 'Product_Category', y = 'Amount')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```
